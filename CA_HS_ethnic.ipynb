```python
import pandas as pd
import numpy as np
from pandas.api.types import is_string_dtype
from pandas.api.types import is_numeric_dtype
from pandas.api.types import is_integer_dtype
```

```python

fresh_ethnic_publicHS = pd.read_csv("/Users/susanlieu/Documents/LIS590OMO/Data/UC/CA_public_HS.csv")
fresh_ethnic_privateHS = pd.read_csv("/Users/susanlieu/Documents/LIS590OMO/Data/UC/CA_private_HS.csv")

counties_from_census = pd.read_csv("/Users/susanlieu/Documents/LIS590OMO/Data/ca_demo.csv")

CA_ethnic_HS_frames = [fresh_ethnic_publicHS, fresh_ethnic_privateHS]
# join the two csv files into one dataframe
CA_HS_ethnic = pd.concat(CA_ethnic_HS_frames)



```

```python
# order by County then Count
#,ascending=False).groupby('job')

CA_HS_ethnic = CA_HS_ethnic.sort_values(["County/State/ Territory", "Count"])
```

```python
# dataframe to get only applicants for each county


app_only = CA_HS_ethnic[(CA_HS_ethnic["Count"] == "App")]
app_only.head()

    
```

```python
#get Applicant count for each ethnicity

app_White = app_only.sort_values(["County/State/ Territory","White"]).groupby("County/State/ Territory").agg({"White":sum})
app_White.reset_index(level=0, inplace=True)
app_Black = app_only.sort_values(["County/State/ Territory","African American"]).groupby("County/State/ Territory").agg({"African American":sum})
app_Black.reset_index(level=0, inplace=True)
app_Native_American = app_only.sort_values(["County/State/ Territory","American Indian"]).groupby("County/State/ Territory").agg({"American Indian":sum})
app_Native_American.reset_index(level=0, inplace=True)                         
app_Latinx = app_only.sort_values(["County/State/ Territory","Hispanic/ Latino"]).groupby("County/State/ Territory").agg({"Hispanic/ Latino":sum})
app_Latinx.reset_index(level=0, inplace=True)
app_Asian = app_only.sort_values(["County/State/ Territory","Asian"]).groupby("County/State/ Territory").agg({"Asian":sum})
app_Asian.reset_index(level=0, inplace=True)
app_other = app_only.sort_values(["County/State/ Territory","Domestic Unknown"]).groupby("County/State/ Territory").agg({"Domestic Unknown":sum})
app_other.reset_index(level=0, inplace=True)

      
```

```python
# dataframe to get Admit count for each county
adm_only = CA_HS_ethnic[(CA_HS_ethnic["Count"] == "Adm")]
adm_only.head()
```

```python
# Get admit count for each ethnicity

adm_White = adm_only.sort_values(["County/State/ Territory","White"]).groupby("County/State/ Territory").agg({"White":sum})
adm_White.reset_index(level=0, inplace=True)
adm_Black = adm_only.sort_values(["County/State/ Territory","African American"]).groupby("County/State/ Territory").agg({"African American":sum})
adm_Black.reset_index(level=0, inplace=True)
adm_Native_American = adm_only.sort_values(["County/State/ Territory","American Indian"]).groupby("County/State/ Territory").agg({"American Indian":sum})
adm_Native_American.reset_index(level=0, inplace=True)
adm_Latinx = adm_only.sort_values(["County/State/ Territory","Hispanic/ Latino"]).groupby("County/State/ Territory").agg({"Hispanic/ Latino":sum})
adm_Latinx.reset_index(level=0, inplace=True)
adm_Asian = adm_only.sort_values(["County/State/ Territory","Asian"]).groupby("County/State/ Territory").agg({"Asian":sum})
adm_Asian.reset_index(level=0, inplace=True)
adm_other = adm_only.sort_values(["County/State/ Territory","Domestic Unknown"]).groupby("County/State/ Territory").agg({"Domestic Unknown":sum})
adm_other.reset_index(level=0, inplace=True)


```

```python
# get full county list from US census
counties = counties_from_census["County"]


# Put Applicants and Admits count into one dataframe for each ethnicity
white = pd.merge(app_White, adm_White, how='outer', on='County/State/ Territory')
white = white.rename(columns={"County/State/ Territory":"County","White_x":"White Applicants", "White_y":"White Admits"})

black = pd.merge(app_Black, adm_Black, how='outer', on='County/State/ Territory')
black = black.rename(columns={"County/State/ Territory":"County","African American_x":"Black Applicants", "African American_y":"Black Admits"})

native = pd.merge(app_Native_American, adm_Native_American, how='outer', on='County/State/ Territory')
native = native.rename(columns={"County/State/ Territory":"County","American Indian_x":"Native American Applicants", "American Indian_y":"Native American Admits"})

latinx  = pd.merge(app_Latinx, adm_Latinx, how='outer', on='County/State/ Territory')
latinx = latinx.rename(columns={"County/State/ Territory":"County","Hispanic/ Latino_x":"Hispanic/Latino Applicants", "Hispanic/ Latino_y":"Hispanic/Latino Admits"})

asian  = pd.merge(app_Asian, adm_Asian, how='outer', on='County/State/ Territory')
asian = asian.rename(columns={"County/State/ Territory":"County","Asian_x":"Asian Applicants", "Asian_y":"Asian Admits"})

other  = pd.merge(app_other, adm_other, how='outer', on='County/State/ Territory')
other = other.rename(columns={"County/State/ Territory":"County","Domestic Unknown_x":"Other Applicants", "Domestic Unknown_y":"Other Admits"})


```

```python
# merge all ethnicity's dataframes into one dataframe, using the full counties list from US census
appl_adm_per_county = pd.DataFrame(counties)

#df1.merge(df2, how='outer').merge(df3, how='outer').merge(df4, how='outer')
appl_adm_per_county = appl_adm_per_county.merge(white, how="outer", on="County").merge(black, how="outer", on="County").merge(native, how="outer", on="County").merge(latinx, how="outer", on="County").merge(asian, how="outer", on="County").merge(other, how="outer", on="County")
appl_adm_per_county


```

```python
# calculate each ethnicity's admittance rate

appl_adm_per_county["White Admit Rate"] = (appl_adm_per_county["White Admits"] / appl_adm_per_county["White Applicants"])*100
appl_adm_per_county["White Admit Rate"] = round(appl_adm_per_county["White Admit Rate"],2)


appl_adm_per_county["Black Admit Rate"] = (appl_adm_per_county["Black Admits"] / appl_adm_per_county["Black Applicants"])*100
appl_adm_per_county["Black Admit Rate"] = round(appl_adm_per_county["Black Admit Rate"],2)

appl_adm_per_county["Native American Admit Rate"] = (appl_adm_per_county["Native American Admits"] / appl_adm_per_county["Native American Applicants"])*100
appl_adm_per_county["Native American Admit Rate"] = round(appl_adm_per_county["Native American Admit Rate"],2)

appl_adm_per_county["Hispanic/Latino Admit Rate"] = (appl_adm_per_county["Hispanic/Latino Admits"] / appl_adm_per_county["Hispanic/Latino Applicants"])*100
appl_adm_per_county["Hispanic/Latino Admit Rate"] = round(appl_adm_per_county["Hispanic/Latino Admit Rate"],2)

appl_adm_per_county["Asian Admit Rate"] = (appl_adm_per_county["Asian Admits"] / appl_adm_per_county["Asian Applicants"])*100
appl_adm_per_county["Asian Admit Rate"] = round(appl_adm_per_county["Asian Admit Rate"],2)

appl_adm_per_county["Other Admit Rate"] = (appl_adm_per_county["Other Admits"] / appl_adm_per_county["Other Applicants"])*100
appl_adm_per_county["Other Admit Rate"] = round(appl_adm_per_county["Other Admit Rate"],2)

```

```python
# get columns as a list

cols = appl_adm_per_county.columns.tolist()
cols
```

```python
# place admit rate after each ethnicity's applicants and admits columns

cols.insert(13, cols.pop(cols.index("Other Admit Rate")))
cols.insert(11, cols.pop(cols.index("Asian Admit Rate")))
cols.insert(9, cols.pop(cols.index("Hispanic/Latino Admit Rate")))
cols.insert(7, cols.pop(cols.index("Native American Admit Rate")))
cols.insert(5, cols.pop(cols.index("Black Admit Rate")))
cols.insert(3, cols.pop(cols.index("White Admit Rate")))

appl_adm_per_county = appl_adm_per_county.reindex(columns= cols)

```
