````python
import csv
with open("/Users/susanlieu/Documents/practicum/Data/zip_codes/uszips_sl.csv", 'r') as fin:
    csvin = csv.reader(fin)
    headers = next(csvin)
    data = [row for row in csvin]
````

````python
headers
````

````python
zipcol = headers.index('zip')
citycol = headers.index('city')
statecol = headers.index('state_name')
countycol = headers.index('county_name')
````

````python
justcali= [row for row in data if row[statecol]=="California"]
````

````python
print(len(justcali))
````

````python
print(justcali)
````

````python
filtered = [ [row[zipcol], row[countycol] ] for row in justcali]
````

````python
print(filtered)
type(filtered)
````

````python
counties = list(set([row[1] for row in filtered]))
````

````python
print(counties)
````

````python
#print(counties)
zips = {county: [] for county in counties} #key: county, value: empty
#print(zips)
#loop thru filtered, use list method to .append() on value; use dict accumlator pattern to add zip code
# end result: {'San Bernadino': ['11111', '22222', '33333']}

for county_zip in filtered:
    zips[county_zip[1]].append(county_zip[0])
    print(zips)



````

````python
# call key to get values
for dict_key in zips:
    print (zips[dict_key])
````
